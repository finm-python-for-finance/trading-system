```python
import yfinance as yf
from pathlib import Path
```

```python
data_dir = Path('data_crypto')
```

```python
tickers = ['BTC-USD', 'ETH-USD', 'SOL-USD']
intervals = ['2m', '5m']
```

```python
for ticker in tickers:
    for interval in intervals:
        data = yf.download(tickers = ticker, interval= interval)
        data.to_csv(data_dir / f"{ticker}_{interval}_data.csv")
```

```python
for ticker in tickers:
    data = yf.download(tickers = ticker, period = "8d", interval = "1m")
    data.to_csv(data_dir / f"{ticker}_1m_data.csv")
```
